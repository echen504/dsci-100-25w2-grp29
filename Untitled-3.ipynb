    Specific Question: "Can a player's level of experience and amount of played hours predict if players are likely to subscribe or not?"

How will the data help me address this question?

```r
library(tidyverse)
library(dplyr)
```

```r
players_data <- read_csv("https://raw.githubusercontent.com/dakyun9/data/refs/heads/main/players%202.csv")
players_data
sessions_data <- read_csv("https://raw.githubusercontent.com/dakyun9/data/refs/heads/main/sessions%202.csv")
sessions_data
```

```r
#tidying all the data 
tidy_players_data <- players_data|>
mutate(gender = as.factor(gender), experience = as.factor(experience))
tidy_players_data

tidy_sessions_data <- sessions_data |>
mutate(start_time = as.factor(start_time), end_time = as.factor(end_time), 
    original_start_time = as.factor(original_start_time), 
    original_end_time = as.factor(original_end_time))
```

```r
#gathering the mean 

full_data <- inner_join(players_data, sessions_data)

average_data<- full_data|>
select(original_start_time, original_end_time, Age, played_hours)|>
summarize(mean_played_hours = mean(played_hours),
          mean_age = mean(Age, na.rm = TRUE), 
          mean_original_start = mean(original_start_time),
          mean_original_end = mean(original_end_time, na.rm=TRUE))
average_data


```

```r
#all the data thats needed for the question

needed_players_data <- players_data |>
select(played_hours, experience, subscribe)
needed_players_data

# needed_sessions_data <- sessions_data|>
# select(start_time, end_time, hashedEmail)

# full_data <- inner_join(needed_players_data, needed_sessions_data)

# tidy_data <- full_data |>
# select(-hashedEmail)
# tidy_data
```

```r
#visualization for all the data needed 

player_graph <- needed_players_data|>
ggplot(aes(x = experience, y = played_hours, color = subscribe))+
geom_point()+
labs(x = "experience level", y = "played amount in hours", color = "subscription status")+
ggtitle("Player's experience and hours played and their subscription status")+
scale_y_log10()
player_graph

#visualization for comparing amounts of subscribtion to experience 

experience_graph <- needed_players_data|>
ggplot(aes(x = subscribe, y = played_hours, color = experience))+
geom_point()+
facet_grid(~ experience)
experience_graph
```

4. Methods and Plan

Method Proposal: The model that I will be using to address my question of interest is the knn classification algorithm. I believe this model is appropriate, as using the k-nearest neighbors will let me predict if an individual as a subscription status or not using played hours and experience levels as variables. I also choose this over regression, because the relationship between the response and predictor is not linear, and will likely underfit the data. Advantages of the classification model is that it requires only a few assumptions, some being that similarity is purely based off of proximity and closeness to other individuals, and that these similar data points are closer to each other. Because of this, there are potential limitations, which include relying on training data to make predictions rather than a generalization that can be applied to other datasets, and may not perform well when the classes are imbalanced. I will process the data by shuffling, stratifying and splitting it into a training set and testing set. This is to ensure that any ordering present in the data doesnt influence it, and also also ensure that around the same proportion of each class ends up on both sets. I will split it in a 25:75 ratio (test and train set respectively). It is also important to scale to remove inconsistencies, missing values and noise.
